## Lets Convert ***Size*** Into Numeric Column

```python
df['Size'].value_counts()
```

```python
def convert_size(size):
    if isinstance(size, float):
        return size
    # size = size.replace(',', '')  # remove comma
    # size = size.replace('+', '')  # remove plus sign
    if size.endswith('M'):
        return float(size[:-1]) * 1024
    elif size.endswith('k'):
        return float(size[:-1])
    elif size == 'Varies with device':
        return np.nan
    else:
        return float(size)

df['Size'] = df['Size'].apply(convert_size)
```

```python
df.sample(2)
```

## ***Install***

```python
df['Installs'].value_counts()
```

```python
def convert_to_numeric(installs):
    if installs == 'Free':
        return 0
    elif installs.endswith('+'):
        installs = installs[:-1]
    installs = installs.replace(',', '')
    return float(installs)

df['Installs'] = df['Installs'].apply(convert_to_numeric)
```

```python
df.head(2)
```

```python
df.describe().head(2)
```

```python
# Lets Do Binnig Of Column Installs to Installs_Group
bins = [0, 10000, 100000, 1000000, 10000000, 100000000, 1000000000]
labels = ['0-10k', '10k-100k', '100k-1M', '1M-10M', '10M-100M', '100M-1B']

#df['Installs'] = df['Installs'].apply(lambda x: x.mid)  # Extract midpoint value from intervals

df['Install_group'] = pd.cut(df['Installs'], bins=bins, labels=labels)
df['Install_group'].value_counts()
```
