<div style="background-color: #f2f2f2; padding: 10px;">
  <h2>Description</h2>
  <p>
    This dataset contains demographic information from the Pakistan Population Census conducted in 2017. It provides detailed population data at various administrative levels within Pakistan, including provinces, divisions, districts, and sub-divisions. The dataset also includes information on urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from the 1998 census, and annual growth rates.
  </p>
</div>

<div style=" padding: 10px;">
  <h2>Features</h2>
  <ul>
    <li>Province: The administrative provinces or regions of Pakistan where the census data was collected.</li>
    <li>Division: The divisions within each province. Divisions are the second level of administrative divisions in Pakistan.</li>
    <li>District: Districts within each division, representing larger administrative units.</li>
    <li>Sub-Division: Sub-divisions or tehsils within each district, providing more localized data.</li>
    <li>Area: The land area (in square kilometers) of each sub-division.</li>
    <li>Urban Population 2017: The population of urban areas within each sub-division for the year 2017.</li>
    <li>Rural Population 2017: The population of rural areas within each sub-division for the year 2017.</li>
    <li>Male Population 2017: The male population within each sub-division for the year 2017.</li>
    <li>Female Population 2017: The female population within each sub-division for the year 2017.</li>
    <li>Transgender Population 2017: The population of transgender individuals within each sub-division for the year 2017.</li>
    <li>Sex Ratio 2017: The sex ratio, calculated as the number of females per 1000 males, within each sub-division for the year 2017.</li>
    <li>Population in 1998: The total population of each sub-division as recorded in the 1998 census.</li>
    <li>Annual Growth Rate: The annual growth rate of the population in each sub-division, calculated as the percentage increase from 1998 to 2017.</li>
  </ul>

````python
print(f"Total Population of Pakistan in Urban Areas Is:{df['ALL SEXES (URBAN)'].sum()}")
````

````python
# How much Rural Population Increased from 1998 to 2017
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()
pop_delta = pop_rural_2017 - pop_rural_1998
print(f"Population Increased From 1998 to 2017 in Rural Areas: {pop_delta}")

# Percentage of Population Increased from 1998 to 2017
per_pop_delta = np.divide(pop_delta, pop_rural_1998) * 100
print(f"Percentage of Population Increased From 1998 to 2017 in Rural Areas: {per_pop_delta}")
````

````python
# How much Rural Population Increased from 1998 to 2017
pop_urban_1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban_2017 = df['ALL SEXES (URBAN)'].sum()
pop_delta = pop_urban_2017 - pop_urban_1998
print(f"Population Increased From 1998 to 2017 in Urban Areas: {pop_delta}")

# Percentage of Population Increased from 1998 to 2017
per_pop_delta = np.divide(pop_delta, pop_urban_1998) * 100
print(f"Percentage of Population Increased From 1998 to 2017 in Urban Areas: {per_pop_delta}")
````
